```python
# Import

import random
import matplotlib.pyplot as plt

marks = [20,35,78,91,10,56,79,16]
grade = [0,0,1,1,0,1,1,0]

w1 = random.random()
epoch = 0
n = 0.01
error_at_epoch = [0]*10

while epoch < 10:
  print("epoch :", epoch, "Weight :", w1)
  i = 0
  while i < 8:
    tr_in = (marks[i])
    #print(tr_in)

    # calculate neuron output

    net = tr_in*w1
    if net>0.5:
      neuron_out = 1
    else:
      neuron_out = 0

    # compare the results

    tr_out = grade[i]
    print(tr_in, tr_out, neuron_out)

    neuron_err = tr_out - neuron_out

    # Adjust the weight

    delta_w = tr_in*w1*neuron_err*n

    w1 = w1 + delta_w

    # Get the absolute error
    error_at_epoch[epoch] = error_at_epoch[epoch] + abs(neuron_err)
    i = i+1

  epoch = epoch+1

plt.plot(error_at_epoch)
plt.show
```
